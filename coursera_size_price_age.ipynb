```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([[1.0, 5.0], [2.0, 10.0]])
y_train = np.array([300.0, 500.0])
m = len(x_train)
```

```python
print(f"x_train : {x_train}")
print(f"y_train : {y_train}")
print(f"m : {m}")
```

```python
i = 0
x_i = x_train[i] # [size, age]
y_i = y_train[i]
print(f"(x^({i})), y^({i})) = ({x_i}, {y_i})")
```

```python
# Create subplots to display both 'size vs price' and 'age vs price'
fig, ax = plt.subplots(1, 2, figsize = (12, 5))

ax[0].scatter(x_train[:, 0], y_train, marker = '*', color = 'r')
ax[0].set_title('House size vs price')
ax[0].set_xlabel('size of house')
ax[0].set_ylabel('price of house')

ax[1].scatter(x_train[:, 1], y_train, marker = 'o', color = 'b')
ax[1].set_title('House age vs price')
ax[1].set_xlabel('age of house')
ax[1].set_ylabel('price of house')

plt.tight_layout()
plt.show()
```

```python
# forrmula : y = w1.x1 + w2.x2 + b
w = np.array([200, -20])
b = 200

print(f"w (poids pour la taille et l'age ) : {w}")
print(f"b (biais) : {b}")
```

```python
#ici on impl√©ment la r√©gression lin√©aire manuellement, sans utiliser scikit-learn ! üöÄ

def compute_model_output(x, w, b):
    """
    Calcule la sortie du mod√®le lin√©aire multiple y = w1*x1 + w2*x2 + ... + b

    X : Matrice des features (taille, √¢ge, etc.)
    w : Vecteur des poids [w1, w2, ...]
    b : Biais (intercept)
    """
    m = x.shape[0]  # recupere le nombre d'√©chantillons
    f_wb = np.zeros(m)  # Initialisation des pr√©dictions (cr√©e un tableau de nbre d'echantilons  o√π chaque √©l√©ment de ce tableau sera remplac√© par une pr√©diction pour chaque maison apr√®s calcul.)

    for i in range(m):
        f_wb[i] = np.dot(x[i], w) + b  # Produit scalaire X * w + b

    return f_wb
    #f_wb est un tableau qui sera utilis√© pour stocker les pr√©dictions du mod√®le pour chaque exemple.
```

```python
y_pred = compute_model_output(x_train, w, b)
print(y_pred)

#Pour la premi√®re maison (taille = 1.0, √¢ge = 5.0), le prix pr√©dit est 250.0.
#Pour la deuxi√®me maison (taille = 2.0, √¢ge = 10.0), le prix pr√©dit est 400.0.
```

```python
# Calcul des pr√©dictions
y_pred = compute_model_output(x_train, w, b)

# Trac√©
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Premier graphique : Taille de la maison vs Prix
ax[0].plot(x_train[:, 0], y_pred, c='b', label='Our prediction')
ax[0].scatter(x_train[:, 0], y_train, marker='*', color='r', label='Actual values')
ax[0].set_title('House size vs price')
ax[0].set_xlabel('Size of house (1000 sqft)')
ax[0].set_ylabel('Price of house (100s of dollars)')
ax[0].legend()

# Deuxi√®me graphique : √Çge de la maison vs Prix
ax[1].plot(x_train[:, 1], y_pred, c='g', label='Our prediction')
ax[1].scatter(x_train[:, 1], y_train, marker='o', color='orange', label='Actual values')
ax[1].set_title('House age vs price')
ax[1].set_xlabel('Age of house (years)')
ax[1].set_ylabel('Price of house (100s of dollars)')
ax[1].legend()

plt.tight_layout()
plt.show()
```
