```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
#m = len(x_train)
m = x_train.shape[0]
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
print(f"Number of training exemple is : {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i})), y^(({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker = 'x', c = 'r')
plt.title('Housing prices')
plt.ylabel('Price in 100s of dollars')
plt.xlabel('Size in 1000 sqft')
plt.show()
```

```python
w = 200
b = 100
print(f"w : {w}")
print(f"b : {b}")
```

```python
def compute_model_output(x, w, b) :
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m) :
    f_wb[i] = w * x[i] + b

  return f_wb
```

```python
y_pred = compute_model_output(x_train, w, b)
plt.plot(x_train, y_pred, c = 'b', label = 'Our prediction')
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Actual values')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
print("Prediction : ", y_pred)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

#Note : In Scikit-Learn, the LinearRegression model expects X to be 2D ((n_samples, n_features))
x_train = np.array([10, 20, 5]).reshape(-1, 1)
y_train = np.array([200, 310, 50])

m = x_train.shape[0]

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

# Predict price for a new house (e.g., size=1.5, age=7)
x_new = np.array([[200]])
y_pred = model.predict(x_new)

y_pred
```
